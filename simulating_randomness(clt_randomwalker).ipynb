```python
# Roll a dice 100 times and plot a histogram of the outcome
import matplotlib.pyplot as plt
import numpy as np
rolls = [random.choice(range(1,7)) for _ in range(100)]
plt.hist(rolls, bins=np.linspace(.5, 6.5, 7))
plt.show()
```

```python
# The intuitively expected result would have been a relatively flat histogram.
# As the laws of large numbers which is a theorem of probablity, that we should
# expect more or less the same number of 1s and 2s all the way through 6s because
# they each have the same probablity.
# Let's try rerunning the simulation with 10,000 repetations.
rolls10k = [random.choice(range(1,7)) for _ in range(10000)]
plt.hist(rolls10k, bins=np.linspace(.5, 6.5, 7))
plt.show()
```

```python
# Not quite flat; let's repeat this simulation for 1 million rolls
rolls1000k = [random.choice(range(1,7)) for _ in range(1000000)]
plt.hist(rolls1000k, bins=np.linspace(.5, 6.5, 7))
plt.show()
```

```python
%%timeit -r1 -n1
# CLT : central limit theorem
# It states that the sum of large number of random variables
# regardless their distribution will approximately follow a 
# normal distribution.
# Test: Sum of 10 dice rolled 1000,000 times
rolls1000k = [sum([random.choice(range(1,7)) for x in range(10)]) for _ in range(1000000)]
plt.hist(rolls1000k)
plt.show()
```

```python
%%timeit -r1 -n1
# Normal Distribution requires the mean and standard deviation
# Mean = 0 , Standard deviation = 1 : Standard normal distribution.
# To generate random numbers form standard normal distribution or from
# the normal distribution in general : 
# np.random.normal(mean, standard devation, length of 1d array or shape(row,column))
# Simulating the above using numpy because it is much faster
# np.random.randint(low, high, shape)
# np.sum(x, axis=dimension)
rolls1000k = np.random.randint(1, 7, (1000000, 10))
sumof10dice1000ktimes = np.sum(rolls1000k, axis=1)
plt.hist(sumof10dice1000ktimes)
plt.show()
```

```python
# Random Walk
# https://en.wikipedia.org/wiki/Random_walk
deltaX = np.random.normal(0, 1, (2,50)) # random walker taking 50 steps
plt.plot(deltaX[0], deltaX[1], "go")
plt.show()
```

```python
# Position of the random walker at any given time
# Cumulative Sum : [1,2,3 .... ] = [1, 1+2, 1+2+3 ...]
X = np.cumsum(deltaX, axis=1)
# Initial location of the random walker
X_0 = np.array([[0], [0]])
newX = np.concatenate((X_0, X), axis=1)
plt.plot(newX[0], newX[1], "go-")
plt.show()
```

```python
# Random walker taking 100 steps
X0 = [[0], [0]]
X1 = np.concatenate((X0,np.cumsum(np.random.normal(0,1,(2,100)), axis=1)), axis=1)
X2 = np.concatenate((X0,np.cumsum(np.random.normal(0,1,(2,100)), axis=1)), axis=1)
X3 = np.concatenate((X0,np.cumsum(np.random.normal(0,1,(2,100)), axis=1)), axis=1)
X4 = np.concatenate((X0,np.cumsum(np.random.normal(0,1,(2,100)), axis=1)), axis=1)
plt.figure()
plt.subplot(221)
plt.plot(X1[0], X1[1], "go-")
plt.subplot(222)
plt.plot(X2[0], X2[1], "go-")
plt.subplot(223)
plt.plot(X3[0], X3[1], "go-")
plt.subplot(224)
plt.plot(X4[0], X4[1], "go-")
plt.show()
```
